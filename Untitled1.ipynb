```python
from PIL import Image
```

```python
import numpy as np
```

```python
image_array = np.zeros((20,200,200,3), dtype=int)
for x in range(1,20):
    temp_image = Image.open(f"{x}.jpg")
    temp_image = temp_image.resize((200,200)) #resize the image
    temp_array = np.array(temp_image)
#     print(temp_array)
#     print(temp_array.shape)
#     print("===============")
    image_array[x] = temp_array
```
